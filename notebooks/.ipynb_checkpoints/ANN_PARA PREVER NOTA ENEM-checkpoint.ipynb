# MODELO DF TUDO SEM QUESTOES FRACAS

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score, classification_report, confusion_matrix
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
X = only_notes.drop('NU_NOTA_MT', axis=1).values
y = only_notes['NU_NOTA_MT'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

# ESCALANDO OS DADOS

```python
scaler = MinMaxScaler()
```

```python
X_train
```

```python
# só fit o train o test não
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

# CRIANDO MODELO

```python
# 27 features ( colunas no modelo de treino)
X_train.shape
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
model = Sequential()

model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))  #neuron nets
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

model.add(Dense(1))  # output

model.compile(optimizer='adam', loss='mse')
```

# TREINANDO O MODELO ESCALADO

```python
# validation data pode mostrar se estamos tendo problemas de overfit, plota o train e test
# batch_size = quanto menor menos chance de overfit(64,128,256...)

model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test),
                                                batch_size=128, epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
predictions = model.predict(X_test)
```

```python
print(mean_absolute_error(y_test, predictions))

```
